```python
%matplotlib inline

import numpy as np
import scipy as sp
import scipy.interpolate as interpolate
import matplotlib.pyplot as plt
from matplotlib.path import Path
import matplotlib.patches as patches
import sys



class Mesh:
    """
    Store the verts and elements and physical data

    attributes
    ----------
    Verts : array
        array of 3d coordinates (npts x 3)
    Elmts : dict
        dictionary of tuples
        (rank 1 array of physical ids, rank 2 array of element to vertex ids
        (Nel x ppe)) each array in the tuple is of length nElmts Phys : dict
        keys and names

    methods
    -------
    read_msh:
        read a 2.2 ascii gmsh file

    write_vtu:
        write VTK file (calling vtk_writer.py)

    notes by XL
        1. This python code assumes the IDs of nodes and elements are continuous 
        and start from 1. However, it is not required in Gmsh document. 
        2. The only element types supported here are: line, triangle, and quadrilateral.
    """
    def __init__(self):

        self.Verts = []
        self.Elmts = {}
        self.Phys = {}

        self.npts = 0
        self.nElmts = {}
        self.nprops = 0
        
        self.hasTri = False
        self.hasQuad = False

        self._elm_types()   # sets elm_type

        self.meshname = ""

    def read_msh(self, mshfile):
        """Read a Gmsh .msh file.

        Reads Gmsh 2.2 mesh files
        """
        self.meshname = mshfile
        try:
            fid = open(mshfile, "r")
        except IOError:
            print "File '%s' not found." % (filename)
            sys.exit()

        line = 'start'
        while line:
            line = fid.readline()

            if line.find('$MeshFormat') == 0:
                line = fid.readline()
                if line.split()[0][0] is not '2':
                    print "wrong gmsh version"
                    sys.exit()
                line = fid.readline()
                if line.find('$EndMeshFormat') != 0:
                    raise ValueError('expecting EndMeshFormat')

            #not so useful for our purpose		
            if line.find('$PhysicalNames') == 0:
                line = fid.readline()
                self.nprops = int(line.split()[0])
                for i in range(0, self.nprops):
                    line = fid.readline()
                    newkey = int(line.split()[0])
                    qstart = line.find('"')+1
                    qend = line.find('"', -1, 0)-1
                    self.Phys[newkey] = line[qstart:qend]
                line = fid.readline()
                if line.find('$EndPhysicalNames') != 0:
                    raise ValueError('expecting EndPhysicalNames')

            if line.find('$Nodes') == 0:
                line = fid.readline()
                self.npts = int(line.split()[0])
                self.Verts = np.zeros((self.npts, 3), dtype=float)
                for i in range(0, self.npts):
                    line = fid.readline()
                    data = line.split()
                    idx = int(data[0])-1  # fix gmsh 1-based indexing
                    if i != idx:
                        raise ValueError('problem with vertex ids')
                    self.Verts[idx, :] = map(float, data[1:])
                line = fid.readline()
                if line.find('$EndNodes') != 0:
                    raise ValueError('expecting EndNodes')

            if line.find('$Elements') == 0:
                line = fid.readline()
                self.nel = int(line.split()[0])
                for i in range(0, self.nel):
                    line = fid.readline()
                    data = line.split()
                    idx = int(data[0])-1  # fix gmsh 1-based indexing
                    if i != idx:
                        raise ValueError('problem with elements ids')
                    etype = int(data[1])           # element type
                    nnodes = self.elm_type[etype]   # lookup number of nodes
                    ntags = int(data[2])           # number of tags following
                    k = 3
                    if ntags > 0:                   # set physical id
                        physid = int(data[k])       # important for the boundary 
                        if physid not in self.Phys:
                            self.Phys[physid] = 'Physical Entity %d' % physid
                            self.nprops += 1
                        k += ntags

                    verts = map(int, data[k:])
                    verts = np.array(verts)-1  # fix gmsh 1-based index

                    if (etype not in self.Elmts) or\
                            (len(self.Elmts[etype]) == 0):
                        # initialize
                        self.Elmts[etype] = (physid, verts)
                        self.nElmts[etype] = 1
                    else:
                        # append
                        self.Elmts[etype] = \
                            (np.hstack((self.Elmts[etype][0], physid)),
                             np.vstack((self.Elmts[etype][1], verts)))
                        self.nElmts[etype] += 1

                line = fid.readline()
                if line.find('$EndElements') != 0:
                    raise ValueError('expecting EndElements')

    
#        print "self.Verts \n"
#        print self.Verts
#        print "self.Elmts "
#        print self.Elmts
#        print "self.Phys \n"
#        print self.Phys

#        print "npts = ", self.npts
#        print "nElmts = ", self.nElmts
#        print "nProps = ", self.nprops	  
    
        fid.close()
        
    def interp_elevation(self, bathyfile):
        """Interpolate the elevation to z cooridnate.

        """
        try:
            fid = open(bathyfile, "r")
        except IOError:
            print "File '%s' not found." % (bathyfile)
            sys.exit()
            
        # Read the number of points
        totalPoints = int(fid.readline())
        print totalPoints
        
        x, y, z = np.linspace(0, 1, totalPoints), np.linspace(0, 1, totalPoints), np.linspace(0, 1, totalPoints)

        # Loop over lines and extract variables of interest
        i=0
        for line in fid:
            line = line.strip()
            columns = line.split()
            x[i] = (columns[0])
            y[i] = (columns[1])
            z[i] = (columns[2])
            i=i+1
            
        fid.close()     
        
        print x.shape
        print y.shape
        print z.shape

        #create the interpolation function
        ifunc = interpolate.interp2d(x, y, z, kind='linear')
              
        xnew = np.linspace(0, 1, self.npts)
        ynew = np.linspace(0, 1, self.npts)

        #get the (x,y) coordiantes of each vertex
        for i in range(0,self.npts):
            xnew[i]=self.Verts[i,0]    
            ynew[i]=self.Verts[i,1]
            
        #interpolate
        znew = ifunc(xnew, ynew)
            
        #set the z coordiante of each vertex
        for i in range(0,self.npts):
            self.Verts[i,2]=znew[i]        

    def write_2dm(self, fname=None):
        """ works for triangles and quadrilaterals in 2d"""

        if fname is None:
            fname = self.meshname.split('.')[0] + '.2dm'

        if type(fname) is str:
            try:
                fid = open(fname, 'w')
            except IOError, (errno, strerror):
                print ".2dm error (%s): %s" % (errno, strerror)
        else:
            raise ValueError('fname is assumed to be a string')

        
        #only support line, triangle, and quadrilateral
        mesh_id_tri = 2   # mesh elements are tris
        mesh_id_quad = 3  # mesh elements are tris
        bc_id = 1   # bdy face elements are lines
        dim = 2

        #elements (cells)
        E_tri = []
        nel_tri = 0
        phy_tri = []
        
        if(mesh_id_tri in self.Elmts):
            self.hasTri = True
            E_tri = self.Elmts[mesh_id_tri][1]
            nel_tri = self.nElmts[mesh_id_tri]
            phy_tri = self.Elmts[mesh_id_tri][0]
            if E_tri.shape[0] != nel_tri:
                raise ValueError('problem with triangle element shape and nel')

                
        E_quad = []
        nel_quad = 0
        phy_quad = []
        if(mesh_id_quad in self.Elmts):
            self.hasQuad = True
            E_quad = self.Elmts[mesh_id_quad][1]
            nel_quad = self.nElmts[mesh_id_quad]
            phy_quad = self.Elmts[mesh_id_quad][0]
            if E_quad.shape[0] != nel_quad:
                raise ValueError('problem with quad element shape and nel')
                
        #print "phy_tri = ", phy_tri        
        #print "phy_quad = ", phy_quad
            
        #boundary (lines)
        Eb = self.Elmts[bc_id][1]
        nelb = self.nElmts[bc_id]
        if Eb.shape[0] != nelb:
            raise ValueError('problem with element shape and nel')

        #boundary IDs 
        bd_id_list = self.Elmts[bc_id][0]
        bd_ids = np.unique(bd_id_list)
        nbc = len(bd_ids)
        
        #print "Eb ", Eb
        #print "nelb ", nelb
        #print "bd_id_list ", bd_id_list
        #print "bd_ids ", bd_ids
        #print "nbc = ", nbc
            

        fid.write('MESH2D\n')
            
        #write the elements (both triangle and quad)    
        for i in range(0, nel_tri):
            data = [i+1]     #ID of the element
            data.extend((E_tri[i, :]+1).tolist())  #Node list of the element
            data.append(phy_tri[i])       #zone (physical group) of the element
            dstr = 'E3T  '       #type of the element
            for d in data:
                dstr += ' %d' % d    
            fid.write(dstr+'\n')   

        for i in range(0, nel_quad):
            data = [i+1+nel_tri]     #ID of the element
            data.extend((E_quad[i, :]+1).tolist())  #Node list of the element
            data.append(phy_quad[i])
            dstr = 'E4Q  '       #type of the element
            for d in data:
                dstr += ' %d' % d    
            fid.write(dstr+'\n')               
            
        #write nodes
        for i in range(0, self.npts):
            fid.write('ND   %d  %e  %e  %e\n' %
                          (i+1, self.Verts[i, 0], self.Verts[i, 1],
                           self.Verts[i,2]))
            
        #write nodestrings
        for i in range(0, nbc):
            cur_BC_ID = bd_ids[i]
            idx_lines=np.nonzero(bd_id_list == cur_BC_ID)    #find the index of all current line segment in Eb
            #print "idx_lines = ", idx_lines[0]
            ele_list = []
            node_list = []
            for idx in idx_lines[0]:
                #print "idx = ", idx+1
                
                if (idx+1) not in ele_list:
                    ele_list.append(idx+1)
                    
                if (Eb[idx][0]+1) not in node_list:
                    node_list.append(Eb[idx][0]+1)
                if (Eb[idx][1]+1) not in node_list:
                    node_list.append(Eb[idx][1]+1)
                    
                    
            #print "node_list = ", node_list        
            #print "ele_list = ", ele_list 
            counter = 0
            fid.write('NS  ')
            for node_I in range(0,len(node_list)):
                if(node_I==(len(node_list)-1)):
                    fid.write('  %d' % -node_list[node_I])
                else:
                    fid.write('  %d' % node_list[node_I])
                
                counter = counter+1
                if(counter==9):   #limit 10 number per line
                    counter = 0
                    fid.write('\n')
                    fid.write('NS  ')
            
            fid.write('\n')
            
            
        
        fid.close()

    def write_vtu(self, fname=None):
        if fname is None:
            fname = self.meshname.split('.')[0] + '.vtu'

        from vtk_writer import write_vtu
        vtk_id = {1: 2, 2: 3, 3: 4}
        Cells = {}
        cdata = {}
        k = 0.0
        for g_id, E in self.Elmts.iteritems():
            print g_id, E
            if g_id in vtk_id:
                Cells[vtk_id[g_id]] = E[1]
                cdata[vtk_id[g_id]] = k*np.ones((E[1].shape[0],))

        write_vtu(Verts=self.Verts, Cells=Cells, cdata=cdata, fname=fname)
  
    def _elm_types(self):
        elm_type = {}
        elm_type[1] = 2    # 2-node line
        elm_type[2] = 3    # 3-node triangle
        elm_type[3] = 4    # 4-node quadrangle

        self.elm_type = elm_type

def plotTri(mesh,colors):
    if mesh.hasTri:
        V=mesh.Verts[:, :2]
        E=mesh.Elmts[2][1]
    
        for i in range(0,len(E)):
            ele = E[i]
            verts = [
               (V[ele[0],0], V[ele[0],1]), 
               (V[ele[1],0], V[ele[1],1]), 
               (V[ele[2],0], V[ele[2],1]), 
               (V[ele[1],0], V[ele[0],1]), ]
            codes = [Path.MOVETO,Path.LINETO,Path.LINETO,Path.CLOSEPOLY,]
            path = Path(verts, codes)
            patch = patches.PathPatch(path, facecolor=colors[mesh.Elmts[2][0][i]], lw=0.1,alpha=0.3)
            plt.gca().add_patch(patch)        
        
def plotQuad(mesh,colors):
    if mesh.hasQuad:
        V=mesh.Verts[:, :2]
        E=mesh.Elmts[3][1]
    
        for i in range(0,len(E)):
            ele = E[i]
            verts = [
               (V[ele[0],0], V[ele[0],1]), 
               (V[ele[1],0], V[ele[1],1]), 
               (V[ele[2],0], V[ele[2],1]), 
               (V[ele[3],0], V[ele[3],1]), 
               (V[ele[1],0], V[ele[0],1]), ]
            codes = [Path.MOVETO,Path.LINETO,Path.LINETO,Path.LINETO,Path.CLOSEPOLY,]
            path = Path(verts, codes)
            patch = patches.PathPatch(path, facecolor=colors[mesh.Elmts[3][0][i]], lw=0.1,alpha=0.3)
            plt.gca().add_patch(patch)

def readmesh(fname,bathy_fname):
    mesh = Mesh()
    mesh.read_msh(fname)

    #interpolate the elevation data; bathy.dat is x y z DEM
    #mesh.interp_elevation(bathy_fname)

    #mesh.write_vtu()
    mesh.write_2dm()
    
    return mesh

if __name__ == '__main__':

    mesh = readmesh('gmsh_straight.msh','bathy.dat')
    
    colors = ['r', 'b', 'g', 'k']
    
    fig = plt.gcf()
    fig.set_size_inches(18.5,10.5)

    plt.ion()
    plotTri(mesh,colors)
    plotQuad(mesh,colors)
    
    plt.axis('scaled')

    plt.show()
```
